```python
import urllib
import json
import time

def get_id(screen_name):
    URL = "https://api.vk.com/method/"
    file = open('/Users/maksimveretennikov/Desktop/Coding/Python/VK_Experiments/access_token.txt', 'r')
    access_token = file.read()
    file.close()
    response = urllib.request.urlopen(URL + 'users.get?user_ids=' + screen_name + '&v=5.80&access_token=' + access_token)
    json_m = response.read().decode('utf-8')
    data = json.loads(json_m)
    id_unknown = '?'
    if 'response' in data:
        id_unknown = str(data['response'][0]['id'])
    elif 'error' in data:
        id_unknown = data['error']['error_msg']
    else:
        id_unknown = "Что-то пошло не так..."
    return id_unknown

def friends_of(user_id):
    URL = "https://api.vk.com/method/"
    file = open('/Users/maksimveretennikov/Desktop/Coding/Python/VK_Experiments/access_token.txt', 'r')
    access_token = file.read()
    file.close()
    #сам запрос
    response = urllib.request.urlopen(URL + 'friends.get?user_id=' + user_id + '&order=hints' + '&fields=nickname' + '&v=5.80&access_token=' + access_token)
    #преобразуем в json формат и в словарь
    json_m = response.read().decode('utf-8')
    data = json.loads(json_m)
    friends_ids = []
    if 'response' in data:
        for i in data['response']['items']:
            friends_ids.append(i['id'])
    elif 'error' in data:
        print(data['error']['error_msg'])
    else:
        print("Что-то пошло не так...")
    return friends_ids

def get_groups(user_id):
    URL = "https://api.vk.com/method/"
    file = open('/Users/maksimveretennikov/Desktop/Coding/Python/VK_Experiments/access_token.txt', 'r')
    access_token = file.read()
    file.close()
    #сам запрос
    response = urllib.request.urlopen(URL + 'groups.get?user_id=' + user_id + "&extended=1" + '&v=5.80&access_token=' + access_token)
    #преобразуем в json формат и в словарь
    json_m = response.read().decode('utf-8')
    data = json.loads(json_m)
    groups_ids = []
    if 'response' in data:
        for i in data['response']['items']:
            groups_ids.append(i['id'])
    elif 'error' in data:
        print(data['error']['error_msg'])
    else:
        print("Что-то пошло не так...")
    return groups_ids

my_id = get_id('for77')
my_friends = friends_of(my_id)
groups_of_friends = {}
for user_id in my_friends:
    groups = get_groups(str(user_id))
    for group in groups:
        if group in groups_of_friends.keys():
            groups_of_friends[group] += 1
        else:
            groups_of_friends.update({group : 1})
    time.sleep(0.3)
print(groups_of_friends)
```
